```python
# -----------------------------------
# User Introduction Program
# -----------------------------------


# Get information from user
print("Can you tell me your name?")
name = input().capitalize()

print("What city do you live in?")
city = input().capitalize()

print("What is you favourite hobby?")
hobby = input()

print("")
# Print a sentence with user information
print(f"Hello {name}! You live in {city} and you enjoy {hobby}.")
```

```python
# -----------------------------------
# Temperature Converter
# -----------------------------------


print("\nEnter temperature value (Celsius): ")
C = float(input())

# Convert Celsius to Fahrenheit
F = (C * 9/5) + 32

print(f"\nTemperature in Fahrenheit is: {F}°F")
```

```python
# -----------------------------------
# Billing Program
# -----------------------------------



# Get information from user
print("\nCan you tell me the Item name?")
name = input()

print("What is the Price of the item?")
price = float(input())

print("what is the Quantity?")
quantity = float(input())

print("")

# Calculate Shoppping Bill
total = price * quantity

# Print item name and total cost
print(f"Total Bill for {name} = {total}")
```

```python
# -----------------------------------
# Student Information Card
# -----------------------------------



# Get information from user/student
print("\nKindly provide your Student Name?")
studentName = input()

print("What is your Class and Section? (in this format: 8A)")
classAndSection = input()

print("What is your School Name?")
schoolName = input()


print("")
print("----- STUDENT ID CARD -----")
print(f"Name: {studentName}")
print(f"Class: {classAndSection}")
print(f"School: {schoolName}")
print("----------------------------")
```
